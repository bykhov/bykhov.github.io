```python
# Decomposition
# Abstraction

def sum_func(a, b):
  """ docstring
  This function returns the sum of two numbers
  Input: a, b
  Output: a + b
  """
  print(a + b)
  result = a + b
  return result, a, b

a,*b = sum_func(1, 2)
print(a,b)
type(a)
sum_func(1,b=2)
```

```python
def is_even(num):
  if num % 2 == 0:
    return True
  else:
    return False

is_even(num=2)
```

```python
# default parameter value
def sum_func(a=1, b=2):
  """ docstring
  This function returns the sum of two numbers
  Input: a, b
  Output: a + b
  """
  result = a + b
  return result

print(sum_func(3,b=2.4))
print(sum_func(3,2.4))
print(sum_func(a=3,b=2.4))
```

```python
def calculator(x,y,op):
  return op(x,y)

def add(x,y):
  return x + y

def sub(x,y):
  return x - y

calculator(1,2,add)
calculator(1,2,sub)
```

##### Variable scope

```python
def my_func(x):
  a = 5
  return x * a

a = 4 # available from the function
my_func(4)
print(a)
```

```python
def my_func(x):
  a += 5
  return x * a

a = 4 # available from the function
my_func(4)
print(a)
```

```python
L = ['a', 'b', 'c']
def my_func(L):
  L.append('d')


my_func(L)
print(L)
```

```python
# function that returns function
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(3)

print(mydoubler(11))
print(mydoubler(22))
myfunc(2)(11)
```
