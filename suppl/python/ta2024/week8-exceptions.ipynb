```python
def divide(x, y):
  '''This function divides two numbers
  Inputs:
      x: int
      y: int
  Output:
      result: float
  '''
  assert y != 0
  assert type(x) == int and type(y) == int, 'Wrong type'
  return x / y

divide(5,1.)
```

```python
def copy(L1,L2):
  ''' Assumes L1,L2 are lists
  Mutates L2 to be a copy of L1
  '''
  while len(L2) > 0:
    L2.pop()
  for e in L1:
    L2.append(e)

L1 = [1,2,3]
L2 = [4,5,6]
copy(L1,L2)
print(L2)
#
```
