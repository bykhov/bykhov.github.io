Write a code that checks if a number, `x`, is non-negative or negative. 
If it is a non-negative number your program should print the value of that variable preceding the text 'is non-negative'.
On the other hand, if it's a negative number, your program should print the value of that variable preceding the text 'is negative'. 
For example, if `x = 1.7` then the program should print: 
'1.7 is non-negative', however if `x = -1.7` then the program should print: '-1.7 is negative'.

```python
x = 12
```

```python
if x >= 0:
  print(f"{x} is non-negative")
  y = 1
else:
  print(f"{x} is negative")
  y = -1
print(y)
```

Write a program that accepts a comma separated sequence of words as input and prints the words in a comma-separated sequence after sorting them alphabetically.

Suppose the following input is supplied to the program:
without,hello,bag,world
Then, the output should be:
bag,hello,without,world

```python
items = [x for x in input().split(',')]
z = items.sort()
print(','.join(items))
print(z)
```

## Differences between the `remove` method and `pop`

```python
students = ["itay", "Roy", "Alon", "Zohar", "Danielle"]
print(students)
x = students.remove("Alon")
print(students, x, sep="\n")
```

```python
students = ["itay", "Roy", "Alon", "Zohar", "Danielle"]
print(students)
x = students.pop(2)
print(students)
y = students.pop(2)
print(students, x, y, sep="\n")
```

## The differences between the **function** `sorted` and the **method** `sort`

```python
lst = [1, 0, -7, 2, 3, -9.1]
print(lst, end="\n\n")
lst_sorted = sorted(lst)
print(lst, lst_sorted, sep="\n")
```

```python
lst = [1, 0, -7, 2, 3, -9.1]
print(lst, end="\n\n")
lst_sorted = lst.sort()
print(lst, lst_sorted, sep="\n")
```

Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number.
The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
values = []
for i in range(1000, 3001):           
    s = str(i)
    if ((int(s[0]) % 2 == 0) and (int(s[1]) % 2 == 0) 
    and (int(s[2]) % 2 == 0) and (int(s[3]) % 2 == 0)):
        values.append(s)
        
print(",".join(values))
```

## `enumerate` and `zip` functions

```python
lst1 = ["a", "b", "c", "d"]
# i = 0
for i, ele in enumerate(lst1): # [(0, "a"), (1, "b"), (2, "c"), (3, "d")]
  print(ele, i)
  # i += 1

# lst1 = ["a", "b", "c", "d"]
# i = 0
# for ele in lst1:
#   print(ele, i)
#   i += 1

print(enumerate(lst1))
print(list(enumerate(lst1)))
```

```python
lst1 = ["a", "b", "c", "d"]
lst2 = [10, 20, 30, 40]
# for ele1 in lst1: 
#   for ele2 in lst2:
#     print(ele1, ele2)

for ele1, ele2 in zip(lst1, lst2): # [("a", 10), ("b", 20), ...]
  print(ele1, ele2)

print(zip(lst1, lst2))
print(list(zip(lst1, lst2)))
```

```python
vector1 = [1, 2, 3]
vector2 = [0, 1, 0]
scalar_prod = 0
for e1 ,e2 in zip(vector1, vector2):
  scalar_prod += e1*e2

print(vector1, vector2, scalar_prod, sep=", ")
```
